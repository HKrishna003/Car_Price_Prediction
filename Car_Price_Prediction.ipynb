## Encoding string class to numeric class

```python
from sklearn.preprocessing import LabelEncoder

# Assuming you have a DataFrame named 'df' containing your dataset

# List of columns you want to encode
columns_to_encode = ['CarName','fueltype','aspiration','doornumber','carbody','drivewheel','enginelocation','enginetype','cylindernumber','fuelsystem']

encoder = LabelEncoder()

# Loop through each column and apply label encoding
for column in columns_to_encode:
    encoded_values = encoder.fit_transform(df[column])
    df[column] = encoded_values
```

```python
df.head()
```

## Seperating dependent and independent columns

```python
X=df.drop(['price'],axis=1)
Y=df['price']
```

## Splitting dataset for training and testing

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,random_state=0)
```
